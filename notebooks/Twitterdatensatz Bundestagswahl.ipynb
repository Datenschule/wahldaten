```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_hashtags = pd.read_csv('../data/hashtags.csv', sep=',')
df_hashtags['Helper'] = 1
```

```python
df_hashtags.text = df_hashtags.text.str.lower()
```

```python
most_hashtags = df_hashtags.groupby(['text']).Helper.agg(['sum']).sort_values(by='sum', ascending=False).head(20).plot(kind='bar', title='Top 10: Most used Hashtags ')
```

```python
plt.show()
```
