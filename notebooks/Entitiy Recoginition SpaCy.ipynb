# Entitiy Recoginition SpaCy

```python
import pandas as pd
import spacy
```

```python
nlp = spacy.load('en')

def entity_rec(entity):
    doc = nlp(str(entity))
    for ent in doc.ents:
        return(ent.label_,ent.text)
```

```python
def check_positions(entry):
    for key, value in category.items():
            if value in str(entry):
                return(key)
```

```python
category = {
            'on advisory board':'on advisory board',
            'Members':'Members',              
            'on payroll':'on payroll', 
            'executive director':'executive director'
            }
```

```python
climate_orgs = pd.read_csv('../data/wiki_excerpt.csv', sep=';')
```

```python
climate_orgs.head(5)
```

```python
climate_orgs['entity'] = climate_orgs.scraped_info.apply(entity_rec)
```

```python
climate_orgs
```

```python
climate_orgs['role'] = climate_orgs.scraped_info.apply(check_positions)
climate_orgs.entity.fillna(climate_orgs.role, inplace=True)
```
